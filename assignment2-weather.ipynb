## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

## Load the Data

```python
url = "https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/main/assignment/weatherreadings1.csv"
df = pd.read_csv(url)

df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])
```

## Plotting:

```python
plt.figure(figsize=(12,8))
plt.plot(df['reportStartDateTime'], df['dryBulbTemperature_Celsius'], marker='o', linestyle='-', color='skyblue', linewidth=2, label='Temperature (C)')

plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature Over Time')

# Format the x-axis to display dates and times
date_formatter = mdates.DateFormatter('%d-%m-%Y %H:%M')
plt.gca().xaxis.set_major_formatter(date_formatter)
plt.gcf().autofmt_xdate()

plt.grid(True)  # Add a grid to the plot
plt.legend()  # Add a legend to the plot
plt.tight_layout()
plt.show()
```
